# Pyber Challenge

### Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_ride_counts_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"].head()
total_ride_counts_by_type
```

```python
# 2. Get the total drivers for each city type
driver_counts_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
driver_counts_by_type
```

```python
#  3. Get the total amount of fares for each city type
fares_total_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
fares_total_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fares_by_type = pyber_data_df.groupby(["type"]).mean()["fare"]
average_fares_by_type
```

```python
# 5. Get the average fare per driver for each city type. 
average_driver_fare = fares_total_by_type / driver_counts_by_type
average_driver_fare
```

```python
#  6. Create a PyBer summary DataFrame. 

pyber_summary_dict = {"Total Rides":total_ride_counts_by_type , 
            "Total Drivers": driver_counts_by_type, 
            "Total Fares": fares_total_by_type,
            "Average Fare per Ride": average_fares_by_type, 
            "Average Fare per Driver": average_driver_fare}


pyber_summary_df = pd.DataFrame(pyber_summary_dict)

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8.  Formatting the DataFrame
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.columns,pyber_data_df.dtypes
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

fares_sum_bydate_df = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
fares_sum_bydate_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fares_sum_bydate_df = fares_sum_bydate_df.reset_index()
fares_sum_bydate_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by_df the date. 
fares_sum_bydate_pivot = fares_sum_bydate_df.pivot(index = "date", columns = "type", values = "fare")
fares_sum_bydate_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
# Note : starter instructions say to use a date range of 1-1-19 to 4-29-19 and instructions in the module have a 
# one day different/earlier stop date of 4-28-19. Results below match the expected results from the module instructions
# despite the difference in dates
jan_apr_fares_df = fares_sum_bydate_pivot.loc['2019-01-01':'2019-04-29']
jan_apr_fares_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8
jan_apr_fares_df.index = pd.to_datetime(jan_apr_fares_df.index)
jan_apr_fares_df.index
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
jan_apr_fares_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
jan_apr_fares_week = jan_apr_fares_df.resample('W').sum()
jan_apr_fares_week
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

#Set up the figure
fig, ax=plt.subplots(figsize = (16,6))
ax.plot(jan_apr_fares_week,label = ["Rural","Suburban","Urban"])
ax.legend(loc='center',title = 'type')
ax.set_title("Total Fare by City Type",fontsize = 20)
ax.set_ylabel("Fare ($USD)")
ax.grid(visible = True)

#Change xticks to months
#code for changing xticks to Months found at 
#https://stackoverflow.com/questions/57586693/how-to-set-x-ticks-to-months-with-set-major-locator

from matplotlib.dates import MonthLocator, DateFormatter
ax.xaxis.set_major_locator(MonthLocator())
ax.xaxis.set_major_formatter(DateFormatter('%b %Y'))

#Save the chart
plt.savefig("analysis/Pyber_Fare_Summary.png")

#Show the plot
plt.show()


```
